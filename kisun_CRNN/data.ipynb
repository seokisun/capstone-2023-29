```python
import pandas as pd
from dateutil import tz
```

```python
df = pd.read_csv('3.20_3.31')
df.info()
df3000 = df[df['route_name'] == '3000']
```

```python
df3000 = df3000.sort_values(by = ['plate_no', 'date_time', 'station_id'])
```

```python
df3000.head(100)
```

```python
df3000.to_csv("timecheck.csv")
```

```python
#Column value type change
df['date_time'] = pd.to_datetime(df['date_time'], format='%Y-%m-%d %H:%M:%S')
df['date_time'] = df['date_time'].dt.tz_convert('Asia/Seoul')
df = df[df['remaining_seats'] != -1]
df.info()
```

```python
df['date_time'] = df['date_time'] + pd.Timedelta(hours=9)
df['DayOfWeek'] = df['date_time'].dt.day_name()
df['day_of_week_int'] = df['date_time'].dt.dayofweek
df['is_weekend'] = (df['day_of_week_int'] >= 5).astype(int)
df.head(1000)
```

```python
df['hour'] = df['date_time'].dt.hour
df['minute'] = df['date_time'].dt.minute//10
df = df.drop(['date_time','plate_no'], axis=1)
```

```python
df = df.drop(['DayOfWeek'], axis=1)
df.head(10)
```

```python
#노선 개수 확인
print(len(df['route_name'].unique()))
#가장 데이터가 많이 쌓인 노선 확인
print(df['route_name'].value_counts().idxmax())
```

```python
#choose most recorded route_name
filtered_df = df[df['route_name'] == '3000']
filtered_df.to_csv("3000_old.csv", index = False)
```

```python
#save modified data
df.to_csv("bus_temp_filterd.csv", index = False)
```
